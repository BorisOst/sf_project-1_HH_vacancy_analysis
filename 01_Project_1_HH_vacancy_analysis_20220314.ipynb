1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
# Копия таблицы с данными
hh_df = hh_data.copy()

def education_transform(x: str):
    """Выделение параметра 'уровень образования' 
    x: str - строка с информацией об образовании
    return: str - возврат уровня образования. Всего 4 уникальных значения:
                [среднее, среднее специальное, неоконченное высшее, высшее]
    """
    # Разделение признака через пробел и выделение первых 2х слов
    x_ls = x.split()[:2]
    
    # Формирование строки для вывода + все слова с маленькой буквы
    if x_ls[1] == 'образование':
        return x_ls[0].lower()
    
    return ' '.join(x_ls).lower()
    

# Формирование признака 'Образование'
hh_df['Образование'] = hh_df['Образование и ВУЗ'].apply(education_transform)

# Проверка преобразования-должно быть 4 уникальных значения+ответ на задание
print(hh_df['Образование'].value_counts())

# Удаление столбца 'Образование и ВУЗ'
hh_df = hh_df.drop(['Образование и ВУЗ'], axis=1)
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def work_experience(x):
    """Выделение опыта работы соискателя
    x: str - строка с опытом работы и информацией о местах работы. 
            Может быть NaN или 'Не указано'
    return: int - опыт работы в месяцах
    """
    # Если информация отсутствует
    if (x == 'Не указано') or (x is np.nan):
        # Возврат nan
        return np.nan
    
    # Списки возможных вариантов написания года и месяца
    year_ls = ['год', 'года', 'лет']
    month_ls = ['месяц', 'месяца', 'месяцев']
    
    # Разделение строки по пробелам и выделение предполагаемого срока
    x_ls = x.split()[2:6]
    
    # Если опыт работы указан в годах и месяцах
    if (x_ls[3] in month_ls) and (x_ls[1] in year_ls):
        # Опыт в месяцах
        month_duration = 12*int(x_ls[0]) + int(x_ls[2])
        
    # Если опыт только в годах
    elif (x_ls[3] not in month_ls) and (x_ls[1] in year_ls):
        # Опыт в месяцах
        month_duration = 12*int(x_ls[0])
        
    # Если опыт только в месяцах
    elif x_ls[1] in month_ls:
        # Опыт в месяцах
        month_duration = int(x_ls[0])
    
    return month_duration
    
# Создание столбца 'Опыт работы (месяц)'
hh_df['Опыт работы (месяц)'] = hh_df['Опыт работы'].apply(work_experience)

# Удаление столбца 'Опыт работы'
hh_df = hh_df.drop(['Опыт работы'], axis=1)

# Медианный опыт работы в месяцах
median_experience = hh_df['Опыт работы (месяц)'].median()
print(f'Медианный опыт в месяцах: {median_experience}')
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
def metro_del(x):
    """Удаляет из строки данные о метро
    x: str - информация о городе, готовности к переезду и командировкам
    return: str - x без информации о метро
    """
    # Удаление информации о метро
    if 'м.' in x:
        # Если информация о метро есть, то:
        # Преобразование в список, разделенный через запятую
        x_modified_lst = x.split(',')
        
        # Удаление информации о метро - в списке №1
        x_modified_lst.remove(x_modified_lst[1])
        
        # Обратное объединение в строку
        x_modified = ','.join(x_modified_lst)
        
        return x_modified
    
    # Если информации о метро нет - возвращает исходную строку
    return x

def brackets_del(x):
    """Удаляет из строки данные в скобках
    x: str - информация о городе, готовности к переезду и командировкам
    return: str - x без данных в скобках
    """
    # Подсчет элементов (в том числе скобок) в строке
    elements_count = Counter(x)
    
    # Очистка строк от информации в скобках
    if elements_count['('] == elements_count[')'] == 0:
        # Если скобок нет - лишнего нет 
        return x
        
    elif elements_count['('] == elements_count[')'] != 0:
        # Если скобки есть и их количество равно - инф в скобках лишняя
        # Объявление модифицируемого элемента
        x_modified = x
        
        # Пока скобки есть:
        while '(' in x_modified:
            # начальный и конечный элементы для удаления
            del_start = x_modified.find('(')
            del_end = x_modified.find(')') + 1
            
            # Если есть вложенные скобки, типа: ...(...(...)...
            if '(' in x_modified[del_start+1:del_end]:
                # Тогда к del_start нужно добавлять delta
                delta = x_modified[del_start:del_end].find('(') + 1
                
                del_start += delta
            
            # Удаление инфомации в скобках
            x_modified = x_modified.replace(
                x_modified[del_start:del_end], ''
            )
        
        return x_modified
    
    elif elements_count['('] != elements_count[')']:
        # Не равное количество - от 1й скобки до конца - лишняя информация
        del_start = x.find('(')
        
        # Удаление элементов от скобки и до конца
        x_modified = x.replace(x[del_start:], '')
        
        return x_modified
    
def str2list(x, split_symbol=','):
    """Преобразование строки в список с удалением лишних пробелов
    x: str - строка с элементами через запятую
    split_symbol=',': str - символ-разделитель
    return: list - список из x с удалением пробелов из элементов в 
            начале и конце элемента
    """
    # Список из строки (разделение по split_symbol)
    x_lst = x.split(split_symbol)
    
    # Удаление лишних пробелов из элементов списка
    x_return_lst = list(map(str.strip, x_lst))
    
    return x_return_lst
    
def needless_del(x):
    """Удаляет из строки данные о метро и информацию в скобках
    x: str - информация о городе, готовности к переезду и командировкам
    return: str - x без лишних данных
    """
    # Удаление информации о метро
    x_modified = metro_del(x)
    
    # Удаление информации в скобках
    x_modified = brackets_del(x_modified)
    
    # Удаление лишних пробелов из очищенной строки и формирование списка
    x_return_lst = str2list(x_modified)
    
    return x_return_lst

# Дополнительный временный столбец с очищенными данными в виде списка
hh_df['Simplify'] = hh_df['Город, переезд, командировки'].apply(needless_del)
```

```python
def city_df(x):
    """Выделение Города
    x: list - информацией о городе, готовности к перезду и командировкам
    return: str - Город: Москва, Питер, миллионник, другой
    """
    # Список городов-миллионников
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 
                      'Казань', 'Челябинск', 'Омск', 'Самара', 
                      'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 
                      'Воронеж', 'Волгоград' ]
    msk_spb = ['Москва', 'Санкт-Петербург']
    
    city = x[0]
    
    if city in msk_spb:
        return city
    
    elif city in million_cities:
        return 'город-миллионник'
    
    return 'другие'

def is_ready2move(x):
    """Выделение готовности к перезду
    x: list - информацией о городе, готовности к перезду и командировкам
    return: bool - True - готов, False - не готов
    """
    if 'не ' in x[1]:
        # Есть отрицание - не готов
        return False
    else:
        return True

def is_ready2trip(x):
    """Выделение готовности к командировкам
    x: list - информацией о городе, готовности к перезду и командировкам
    return: bool - True - готов, False - не готов
    """
    # В некоторых строках есть информация только о готовности к переезду
    # В этих случаях длина x равна 2
    try:
        if 'не ' in x[2]:
            # Есть отрицание - не готов
            return False
        else:
            return True
    except:
        # Если нет информации о командировках
        return False


# Формироавние признака Город
hh_df['Город'] = hh_df['Simplify'].apply(city_df)

# Формирование признака Готовность к переезду
hh_df['Готовность к переезду'] = hh_df['Simplify'].apply(is_ready2move)

# Формирование признака Готовность к командировкам
hh_df['Готовность к командировкам'] = hh_df['Simplify'].apply(is_ready2trip)

# Удаление ненужных столбцов: "Город, переезд, командировки" и "Simplify"
hh_df.drop(['Город, переезд, командировки', 'Simplify'], axis=1, inplace=True)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость

* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
def column2list(x):
    """НУЖНА ТОЛЬКО ДЛЯ СПОСОБА №1
    Из списка (столбец DataFrame) делает список уникальных значений
    x: list - список со списками признаков 
    return: list - список уникальных значений признаков
    """
    # Инициализация множества
    return_set = set()
    
    # Преобразование списков в множества и их объединение
    for element_lst in x:
        return_set = return_set.union(set(element_lst))
        
    return list(return_set)


# Извлечение из ст-ца "Занятость" списков. Ф-я str2list из предыдущего задания
hh_df['Занятость'] = hh_df['Занятость'].apply(str2list)

# То же для столбца График
hh_df['График'] = hh_df['График'].apply(str2list)

##################################################
# Способ 1. Таким способом решение получается, но не быстро (~1min 22s)
# Список уникальных значений для 2х столбцов
# employment_uniq = column2list(list(hh_df['Занятость']))
# schedule_uniq = column2list(list(hh_df['График']))

# Добавление столбцов с уникальными значениями из Занятость и График
# hh_df[employment_uniq] = np.nan
# hh_df[schedule_uniq] = np.nan

# Заполнение столбцов с уникальными значениями
# for i in range(hh_df.shape[0]):
#     for elem in hh_df.loc[i, 'Занятость']:
#         hh_df.loc[i, elem] = True
#     for elem in hh_df.loc[i, 'График']:
#         hh_df.loc[i, elem] = True

# Заполнение NaN значениями False
# sum_lst = employment_uniq[:]
# sum_lst.extend(schedule_uniq)
# hh_df[sum_lst].fillna(value=False)

##################################################
# Способ 2. Принцип подсмотрен на Stackoverflow:
# https://stackoverflow.com/questions/29034928/pandas-convert-a-column-of-list-to-dummies
# Гораздо быстрее (~12.8 s) и симпатичней
# Создается два DF с расписанными признаками для 2х столбцов
df_employment = pd.get_dummies(
    hh_df['Занятость'].apply(pd.Series).stack()
).sum(level=0).astype(bool)

df_schedule = pd.get_dummies(
    hh_df['График'].apply(pd.Series).stack()
).sum(level=0).astype(bool)

# Объединение в один DF
hh_df = pd.concat([hh_df, df_employment, df_schedule], axis=1)

##################################################
# ЗАДАНИЕ 3.5
# Маски - проектная работа и волонтерство, вахтовый метод и гибкий график
mask_01 = hh_df['проектная работа'] & hh_df['волонтерство']
mask_02 = hh_df['вахтовый метод'] & hh_df['гибкий график']

# Ответы на задание
print(f'Ищут проектная работа и волонтерство: {hh_df[mask_01].shape[0]}')
print(f'Ищут вахтовый метод и гибкий график: {hh_df[mask_02].shape[0]}')

# Удаление столбцов Занятость и График
hh_df.drop(['Занятость', 'График'], axis=1, inplace=True)
```

Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# Загрузка курсов валют 
exchange_rates = pd.read_csv('data/ExchangeRates.csv', sep=',')

# Преобразование столбца Обновление резюме в дату - двойное преобразование -
# чтобы получить чтобы получить дату в нужном формате
hh_df['Обновление резюме'] = pd.to_datetime(pd.to_datetime(
    hh_df['Обновление резюме'], format="%d.%m.%Y %H:%M"
).dt.date)

# То же для даты в таблице курса валют
exchange_rates['date'] = pd.to_datetime(
    exchange_rates['date'], format="%d/%m/%y"
)

# Преобразование строк в список. Функция - из задания 4
hh_df['ЗП'] = hh_df['ЗП'].apply(str2list, split_symbol=' ')

# Формирование 2х временных стобцов с величиной ЗП и валютой (не ISO)
hh_df['ЗП, сумма'] = hh_df['ЗП'].apply(lambda x: float(x[0]))
hh_df['ЗП, валюта'] = hh_df['ЗП'].apply(lambda x: x[1])

# Преобразование валюты в ISO-формат
currency2iso = {'руб.': 'RUB', 'KZT': 'KZT', 
                'USD': 'USD','бел.руб.': 'BYN',
                'EUR': 'EUR','грн.': 'UAH',
                'сум': 'UZS','KGS': 'KGS',
                'AZN': 'AZN'}

hh_df['ЗП, ISO'] = hh_df['ЗП, валюта'].apply(lambda x: currency2iso[x])

# Приоединение к таблице с резюме таблицы с курсами по столбцам Дата и Валюта
hh_df = hh_df.merge(
    exchange_rates[['date', 'currency', 'close', 'proportion']], 
    left_on= ['Обновление резюме', 'ЗП, ISO'],
    right_on= ['date', 'currency'],
    how='left'
)

# Обработка NaN в столбцах close и proportion. Заполнение 1, ибо рубль
hh_df[['close', 'proportion']] = hh_df[['close', 'proportion']].fillna(1.0)

# Столбец с зарплатой в рублях
hh_df['ЗП (руб)'] = hh_df['ЗП, сумма'] * hh_df['close'] / hh_df['proportion']

# Удаление лишних столбцов
columns4del = ['ЗП', 'ЗП, валюта', 'date', 'currency', 
               'ЗП, сумма', 'ЗП, ISO', 'close', 'proportion']
hh_df.drop(columns4del, axis=1, inplace=True)

##################################################
# Задание 3.6
# Медианная зарплата в тысячах рублей
median_salary = round(1e-3 * hh_df['ЗП (руб)'].median())

print(f'Медианная зарплата, тыс.рублей: {median_salary}')
```

7. Постройте сводную таблицу, иллюстрирующую зависимость **медианной** желаемой заработной платы от возраста (**"Возраст"**) и образования (**"Образование"**). На полученной сводной таблице постройте **тепловую карту**. Проанализируйте тепловую карту, сравнив показатели внутри групп.

```python
def get_age_cat(x):
    """Преобразование возраста в категориальный признак
    x: int - возраст, число 
    return: str - строка - категория возраста
    """
    if x <= 20:
        return '<=20'
    elif 20 < x <= 30:
        return '20<x<=30'
    elif 30<x<=40:
        return '30<x<=40'
    elif 40<x<=50:
        return '40<x<=50'
    elif 50<x<=60:
        return '50<x<=60'
    elif 60<x<=70:
        return '60<x<=70'
    elif 70<x:
        return '70<x'

    
# Преобразование возраста в категориальный признак
hh_df['Возраст_cat'] = hh_df['Возраст'].apply(get_age_cat)

# Сводная таблица ЗП от возаста и образования
pivot_salary = hh_df.pivot_table(values='ЗП (руб)',
                                 index='Образование',
                                 columns='Возраст_cat',
                                 aggfunc='median'
                                )

# Изменение порядка столбцов
pivot_salary = pivot_salary[['<=20', '20<x<=30', '30<x<=40', 
              '40<x<=50', '50<x<=60', 
              '60<x<=70', '70<x']]

# Видоизмененная тепловая карта
sns.set_theme()

# Построение графика
f, ax = plt.subplots(figsize=(12, 8))
sns.heatmap(
    data=pivot_salary, 
    annot=True, 
    linewidths=0.5, 
    ax=ax
);

ax.set_title('Медианная ЗП от Возраста и Образования', 
             fontdict={'fontsize': 16});
ax.set_xlabel('Возраст', fontdict={'fontsize': 14})
ax.set_ylabel('Образование', fontdict={'fontsize': 14})

# Удаление столбца Возраст_cat
hh_df = hh_df.drop(['Возраст_cat'], axis=1)
```

Для всех категорий образования при возрасте соискателя <=20 лет медианная зарплата одинакова. С увеличением возраста соискателя начинается рост ЗП, но неравномерный. Например, для людей с высшим образованием рост наиболее быстрый и к 40-50 годам достигает максимального значения (рост более чем в 2.5 раза), после чего наблюдается некоторый спад, с последующим некоторым восстановлением после 70 лет. 

Для категории соискателей с неоконченным высшим образованием также сначала наблюдается некоторый рост медианной ЗП (максимум 40-50 лет), после чего стремительное падение до величин (60-70лет), меньших, чем в самом младшем возрастем (<20 лет). Для возраста >70 лет данных нет.

Более равномерно ведут себя значения медианной ЗП для категории среднего  образования - величина медианной ЗП меняется не более чем в 1.5 раза. Рост зарплты наиболее медленный, достигает максимума для возраста 30-50лет. Для лиц старше 60 лет данные отсутствуют.

Для соискателей со средне специальным образованием рост зарплаты небыстрый и небольшой, не более чем в 1.5 раза для возраста 30-40лет. Величина остается практически постоянной до ~60 лет, с последующим падением для 60-70 лет и стремительным ростом (почти в 2 раза) после 70лет.

Таким образом. Максимальное трудовое долголетие наблюдается у людей с высшим и средне специальным образованием. Максимальные зарплатные ожидания у людей с высшим образованием в возрасте 40-50лет. Наименьшая ЗП - неоконченное высшее после 60 лет. Наиболее стабильный уровень ЗП у соискателей со средне специальным образованием, резкий рост после 70 лет

```python
# Сначала целесообразно провести заполнение пропусков столбца Опыт медианами
value = {'Опыт работы (месяц)': hh_dedupl['Опыт работы (месяц)'].median()}
hh_fill = hh_dedupl.fillna(value)

# Количество пропусков по столбцам - убедиться, что в Опыте все норм
hh_fill.isnull().sum()
# Все норм!

# Удаление строк с пропусками
hh_drop = hh_fill.dropna(how='any', axis=0)
print(f'Итоговая длина DF после удаления пропусков: {hh_drop.shape[0]}')

# Количество пропусков по столбцам
hh_drop.isnull().sum()
# Все норм!

# Результирующее среднее значение Опыта
experience_mean = round(hh_drop['Опыт работы (месяц)'].mean())
print(f'Среднее значение опыта в месяцах: {experience_mean}')
```

4. Мы добрались до ликвидации выбросов. Сначала очистим данные вручную. Удалите резюме, в которых указана заработная плата либо выше 1 млн. рублей, либо ниже 1 тыс. рублей.

```python
# Маска для ЗП больше миллиона и меньше тысячи
mask_outliers = (hh_drop['ЗП (руб)'] < 1e3) | (hh_drop['ЗП (руб)'] > 1e6)

print(f'Количество выбросов (ЗП > 1e6 или < 1e3): {mask_outliers.sum()}')

# Удаление выбросов с ЗП > 1e6 или < 1e3
hh_del_outliers_step1 = hh_drop.drop(
    hh_drop[mask_outliers].index, 
    axis=0)

print(f'Длина DF после удаления экстремальных ЗП: {hh_del_outliers_step1.shape[0]}')
```

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
def outliers_z_score_mod(data, left=3, right=3):
    """Реализует метод сигм
    Код функции частично взят из Курса, раздел по Очистке данных
    data: Series - данные, для которых реализуется поиск выбросов методом сигм
    left: int - число сигм для отступа влево от матожидания
    right: int - число сигм для отступа вправо от матожидания
    return: outliers, cleaned: Series - выбросы и Series без выбросов
    """
    mu = data.mean()
    sigma = data.std()
    lower_bound = mu - left*sigma
    upper_bound = mu + right*sigma
    
    mask_out = (data < lower_bound) | (data > upper_bound)
    mask_clean = (data > lower_bound) & (data < upper_bound)
    
    outliers = data[mask_out]
    cleaned = data[mask_clean]
    
    return outliers, cleaned


# Series с логарифмом возраста
log_age = np.log(hh_del_outliers_step2['Возраст'])

# Матожидание и стандартное отклонение логарифма возраста
log_age_mean = log_age.mean()
log_age_std = log_age.std()

# Нижняя и верхняя границы метода z-отклонений
lower_bound = log_age_mean - 3*log_age_std
upper_bound = log_age_mean + 3*log_age_std

# Построение графика Возраста в логарифмическом масштабе
fig = plt.figure(figsize=(10, 7))

# Гистограмма распределения возраста соискателей в log масштабе
histplot = sns.histplot(
    data=log_age, 
    bins=55, 
    kde=True, 
    color='blue'
);

histplot.axvline(log_age_mean, color='k', lw=2)
histplot.axvline(lower_bound, color='r', lw=2, linestyle='--')
histplot.axvline(upper_bound, color='r', lw=2, linestyle='--')
histplot.axvline(
    log_age_mean + 4*log_age_std, color='g', lw=2, linestyle='--'
)

histplot.set_title('Распределение log(возраста) соискателей', 
                   fontdict={'fontsize': 16})
histplot.set_xlabel('log(Возраст)', 
                   fontdict={'fontsize': 14})
histplot.set_ylabel('Count', 
                    fontdict={'fontsize': 14});

print(
    f'Асимметрия расп-я правосторонняя: As={round(log_age.skew(), 2)} > 0'
)

# Series с выбросами и очищенными данными
hh_outliers, hh_cleaned = outliers_z_score_mod(log_age, right=4)

print(f'Длина полностью очищенного DF: {hh_cleaned.shape[0]}')

# Таблица с полученными выбросами
hh_del_outliers_step2.loc[hh_outliers.index]
```
